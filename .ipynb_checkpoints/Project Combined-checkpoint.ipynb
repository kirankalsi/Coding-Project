```python
url = ("https://www.inflation.eu/inflation-rates/great-britain/historic-inflation/cpi-inflation-great-britain.aspx")
page = requests.get(url).text  # make the request from url
soup = BeautifulSoup(page, "html.parser")  # pull the entire HTML file

link = soup.find_all('table')  # get all tables
tables = link[14]  # get 15th table 
trs = tables.find_all('tr')  # get all trs in that table
inflationRate = []  # initialise inflation rate array

for index,ele in enumerate(trs):  # loop through each element in trs
    tds = trs[index].find_all("td")  # get all tds in each tr
    infRate = (tds[len(tds)-1]).text  # get last td text from tds
    infRate2 = (tds[1]).text  # # get second td text from tds
    inflationRate.append({'date': (tds[0]).text ,'value': literal_eval("'%s'" % infRate2)})  # convert numbers into text and obtain actual values
    inflationRate.append({'date': (tds[len(tds)-2]).text ,'value': literal_eval("'%s'" % infRate)})  # convert numbers into text and obtain actual values
print(inflationRate)
```

```python
CPI = pd.DataFrame(inflationRate)  # convert array to dataframe
CPI = CPI.drop([0,1])  # drop rows
CPI["date"] = CPI["date"].str.replace(r'\D+', '')  # keep digits in date column
CPI.head()
```

```python
def scatter_plot(x1name, x2name, title):  # define scatter plot function
    x1 = Variables_df[x1name]  # look up independent variable 1
    x2 = Variables_df[x2name]  # look up independent variable 2
    y = Variables_df['totalEvCarSales']  # dependent variable

    fig = plt.figure(figsize=(20, 8))  # create figure

    ax1 = fig.add_subplot(121) # set up axes with 1 row and 2 columns, first plot
    ax2 = fig.add_subplot(122) # second plot

    ax1.scatter(x1, y1, color='#0000ff')  # plot data
    ax1.set(xlabel=x1name, ylabel="Total Electric Car Sales")  # set labels

    ax2.scatter(x2, y1, color='#ff0000')  # plot data
    ax2.set(xlabel=x2name, ylabel="Total Electric Car Sales")  # set labels

    fig.suptitle(title, fontsize=24)  # set title for the whole figure

    plt.show()  # show the plot
```

```python
scatter_plot('Charging Connectors', 'Bus Fare Cost', 'Strong Positive Correlations')  # call function with independent variables and title
```

```python
scatter_plot('Carbon Monoxide (thousand tonnes)', 'Electricity Available (TWh)', 'Strong Negative Correlations')  # call function with independent variables and title
```

```python
def dot_whisk_plot(model):  # define dot-whisker plot function
    series = model.params - model.conf_int()[0]  # calcualte the error bar lengths for confidence intervals
    coef_df = pd.DataFrame({'Coefficient': model.params.values[1:],  # create coefficient dataframe
                        'Error Term': series.values[1:],
                        'Variable': series.index.values[1:]
                       })
    
    fig, ax = plt.subplots(figsize=(14, 6))  # create figure
    coef_df.plot(x='Variable', y='Coefficient', kind='bar',  # plot bar charts
                 ax=ax, color='blue', 
                 yerr='Error Term', legend=False)
    ax.scatter(x=pd.np.arange(coef_df.shape[0]),  # plot scatter graph
               marker='s', s=120, 
               y=coef_df['Coefficient'], color='black')
    ax.axhline(y=0, linestyle='--', color='red', linewidth=1)  # plot y=0 dashed line
```

```python
dot_whisk_plot(reg_model1_fit)  # call function with fitted model
```
